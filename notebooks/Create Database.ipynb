# Games

```python
season = pd.read_csv('../data/historic/regular_season_detailed_results.csv')
season_2015 = pd.read_csv('../data/2015/regular_season_detailed_results_2015.csv')
season = pd.concat((season, season_2015))#print season.tail()

season['lucky'] = np.maximum(0, 5 - season.wscore + season.lscore)

# APBR metrics
season['wposs'] = 0.96 * (season.wfga - season.wor + season.wto + 0.475 * season.wfta)
season['lposs'] = 0.96 * (season.lfga - season.lor + season.lto + 0.475 * season.lfta)

season['wrat'] = season.wscore / season.wposs
season['lrat'] = season.lscore / season.lposs

season['wefg'] = (season.wfgm + 0.5 * season.wfgm3) / season.wfga
season['lefg'] = (season.lfgm + 0.5 * season.lfgm3) / season.lfga

season.head(3)
```

```python
season.to_sql('games', engine, if_exists='replace', index=False)
```

# Team stats

```python
class PageRank:
    def __init__(self, rho, iters, f, colname='pagerank'):
        self.rho = rho
        self.iters = iters
        self.prs = {}
        self.f = f
        self.colname = colname
        
    def compute(self, games, teams):
        for year in games.season.unique():
            self.compute_year(games[games.season == year], year)
        
        teams[self.colname] = teams.apply(self.pd_pr, 1)
    
    def create_A(self, data, n, to_idx):
        A = self.rho * np.ones((n, n))
        
        for idx, row in data.iterrows():
            A[to_idx[row.wteam], to_idx[row.lteam]] += self.f(row)
        return A
                
    def compute_pr(self, A):
        n = A.shape[0]
        v = np.ones(n) 
        for _ in xrange(self.iters):
            v = v / np.sqrt(np.dot(v, v))
            v = np.dot(A, v)
        v = v / np.sqrt(np.dot(v, v))  
        return v * n
    
    def compute_year(self, data, year):
        unique_ids = pd.concat((season.wteam, season.lteam)).unique()
        to_idx = {x: i for i,x in enumerate(unique_ids)}
        n = len(to_idx)
        A = self.create_A(data, n, to_idx)
        v = self.compute_pr(A)
        self.prs[year] = {k: v[to_idx[k]] for k in unique_ids}
        
    def get_pr(self, team, season):
        try:
            return self.prs[int(season)][int(team)]
        except KeyError:
            print 'Team: {}, season: {} not found'.format(team, season)
            
    def pd_pr(self, row):
        return self.get_pr(row.team_id, row.season)
    
```

```python
# create game stats
def swap_loc(row):
    if row.wloc == 'H':
        return 'A'
    elif row.wloc == 'A':
        return 'H'
    return 'N'


t_wins = season.rename(columns={'wteam':'team_id', 'wscore': 'score', 
                           'lteam':'opponent_id', 'lscore':'opp_score'})
t_losses = season.rename(columns={'wteam':'opponent_id', 'wscore': 'opp_score', 
                           'lteam':'team_id', 'lscore':'score'})

t_wins.rename(columns={x: x[1:] for x in season.columns[8:21]}, inplace=True)
t_wins.rename(columns={x: 'opp_{}'.format(x[1:]) for x in season.columns[21:34]}, inplace=True)
t_wins.rename(columns={'wposs': 'poss', 'lposs': 'opp_poss', 'wrat': 'oer', 'lrat': 'der',
                       'wefg': 'efg', 'lefg': 'opp_efg'}, inplace=True)
t_wins['loc'] = t_wins.wloc
t_wins.drop('wloc', 1, inplace=True)
t_losses.rename(columns={x: 'opp_{}'.format(x[1:]) for x in season.columns[8:21]}, inplace=True)
t_losses.rename(columns={x: x[1:] for x in season.columns[21:34]}, inplace=True)
t_losses.rename(columns={'lposs': 'poss', 'wposs': 'opp_poss', 'lrat': 'oer', 'wrat': 'der',
                       'lefg': 'efg', 'wefg': 'opp_efg'}, inplace=True)
t_losses['loc'] = t_losses.apply(swap_loc, 1)
t_losses.drop('wloc', 1, inplace=True)
# t_losses.lucky = -t_losses.lucky

t_wins['outcome'] = 1
t_losses['outcome'] = 0

games = pd.concat((t_wins, t_losses))
games.drop('lucky', 1, inplace=True)
games.head()
```

```python
# fix teams to not use games
teams = games.groupby(['team_id', 'season']).agg({'outcome': 'sum', 'opponent_id': 'count', 
                                               'score': 'mean', 'opp_score': 'mean', 'efg': 'mean',
                                               'opp_efg': 'mean', 'poss': 'mean', 'opp_poss': 'mean',
                                               'oer': 'mean', 'der': 'mean'}).reset_index()
teams.rename(columns={'outcome': 'wins', 'opponent_id': 'ngames'}, inplace=True)


lucky_wins = season.groupby(['wteam', 'season']).agg({'lucky': 'sum'}).rename(columns={'lucky': 'wluck'}).reset_index()
unlucky_losses = season.groupby(['lteam', 'season']).agg({'lucky': 'sum'}).rename(columns={'lucky': 'lluck'}).reset_index()
luck = pd.merge(lucky_wins, unlucky_losses, how='outer', left_on=['season', 'wteam'], right_on=['season','lteam']).drop('lteam', 1).rename(columns={'wteam': 'team_id'})
luck.fillna(0, inplace=True) 
luck['luck'] = luck.wluck - luck.lluck
luck.drop(['wluck', 'lluck'], 1, inplace=True)

teams = pd.merge(teams, luck, on=['team_id', 'season'])


f = lambda row: np.sqrt(row.wscore - row.lscore + 5 + 3 * (row.wloc != 'H'))
pr = PageRank(0.05, 30, f, 'winrank')
pr.compute(season, teams)

teams['teamrank'] = teams.groupby(['season'])['winrank'].rank(ascending=False).astype(int)

# teams['win_perc'] = teams['wins'] / teams['ngames']
# teams['record_m'] = teams['wins_m'] / teams['ngames_m']
# teams['avg_margin_m'] = teams['score_m'] - teams['opp_score_m']

teams.head(3)
```

```python
teams.to_sql('team_stats', engine, if_exists='replace', index=False, index_label='team_id')
```
